```python
import requests
import pandas as pd

# Token da API
token = "cb6a3ddc-32f0-4f14-b969-5c03dfd45577"

# URL base da API
url = "https://api.gupy.io/api/v2/candidates"

# Headers com o token
headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {token}"
}

# Lista para armazenar todos os resultados
all_results = []

# Inicializa o nextPageToken
next_page_token = None

# Define as datas para filtrar outubro de 2024
updated_after = "2024-10-01T00:00:00Z"
updated_before = "2024-10-15T23:59:59Z"

while True:
    # Parâmetros da requisição
    params = {
        "perPage": 100,  # Defição da quantidade de resultados por página
        "updatedAfter": updated_after,  # Filtro para obter resultados após 1º de outubro
        "updatedBefore": updated_before  # Filtro para obter resultados antes de 15 de outubro
    }
    if next_page_token:
        params["nextPageToken"] = next_page_token  # Adiciona o token se existir

    # Requisição GET
    response = requests.get(url, headers=headers, params=params)

    # Verificando se a requisição foi bem-sucedida
    if response.status_code == 200:
        data = response.json()

        # Imprimindo a resposta completa para depuração
        #print("Resposta da API:", data)

        # Acessando os dados na coluna 'results'
        if 'results' in data:
            results_data = data['results']
            all_results.extend(results_data)  # Adiciona os resultados à lista

            # Atualiza o nextPageToken para a próxima requisição
            next_page_token = data.get('nextPageToken')

            # Log de status
            print(f"Fetched {len(results_data)} results, nextPageToken: {next_page_token}")

            # Exibindo a tabela completa da página atual
            df_partial = pd.DataFrame(results_data)
            print("Resultados da página atual:")
            print(df_partial)  # Exibe toda a tabela

            # Verifique se o nextPageToken é None ou vazio
            if not next_page_token:
                print("Não há mais páginas para buscar.")
                break
        else:
            print("A coluna 'results' não foi encontrada nos dados.")
            break
    else:
        print(f"Erro: {response.status_code}, {response.text}")
        break

# Transformando todos os resultados em um DataFrame do pandas
df = pd.DataFrame(all_results)

# Exibindo sobrenomes únicos de outubro de 2024
print("Sobrenomes únicos encontrados:")
print(df['lastName'].unique())

# Exibindo a tabela final
print("Tabela final com todos os resultados:")
print(df)
```
